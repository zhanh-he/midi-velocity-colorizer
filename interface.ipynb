```python
import numpy as np
import matplotlib.pyplot as plt

# Path to your npz file containing attention weights
attn_path = "results/npy_result/UNet_attention_1.npy.npz"  # Update with your actual path
attn_npz = np.load(attn_path)

print("Attention keys:", attn_npz.files)

for key in attn_npz.files:
    attn_array = attn_npz[key]
    print(f"Original shape for '{key}':", attn_array.shape)
    
    # Process the attention array for visualization.
    # For instance, if the attention array is 3D with an inhomogeneous channel, 
    # take the mean over the last dimension if necessary.
    if attn_array.ndim == 3 and attn_array.shape[-1] not in (1, 3, 4):
        attn_to_plot = attn_array.mean(axis=-1)
        print(f"Reduced shape for '{key}' after averaging over last dimension:", attn_to_plot.shape)
    elif attn_array.ndim == 4:
        # Select first batch element, then reduce if necessary.
        attn_to_plot = attn_array[0]
        if attn_to_plot.ndim == 3 and attn_to_plot.shape[-1] not in (1, 3, 4):
            attn_to_plot = attn_to_plot.mean(axis=-1)
            print(f"Reduced shape for '{key}' after batch selection and averaging:", attn_to_plot.shape)
    else:
        attn_to_plot = attn_array
    
    plt.figure(figsize=(5, 4))
    plt.imshow(attn_to_plot, aspect='auto', cmap='viridis')
    plt.title(f"Attention Weights for '{key}'")
    plt.xlabel("Dimension 2")
    plt.ylabel("Dimension 1")
    plt.gca().invert_yaxis() 
    plt.colorbar(label="Attention Value")
    plt.show()
```

```python
import torch
from pprint import pprint
import matplotlib.pyplot as plt

# Set the path to your checkpoint file
ckpt_path = "results/checkpoints_bk/SMD-UNet-t10-attn2 normal2-BCELoss maskTrue u_shape cosim0.2-bat3-kaiming-element_wise-unsw2 A6000/epoch=299.ckpt"

# Load the checkpoint
# Use map_location='cpu' to load it on CPU (change as needed)
checkpoint = torch.load(ckpt_path, map_location='cpu')

# Print out the keys of the checkpoint to see what it contains
print("Checkpoint keys:")
pprint(list(checkpoint.keys()))
# You will typically see keys like 'state_dict', 'epoch', 'callbacks', etc.

# Now inspect the state_dict (which contains model parameters)
state_dict = checkpoint.get("state_dict", None)
if state_dict is None:
    print("No state_dict found in checkpoint!")
else:
    print("\nState_dict keys:")
    pprint(list(state_dict.keys()))
    
    # For example, print out the shape of each parameter tensor:
    print("\nParameter shapes:")
    for key, tensor in state_dict.items():
        if isinstance(tensor, torch.Tensor):
            print(f"{key}: {tensor.shape}")

    # Optionally, visualize a weight distribution from one of the layers:
    # Let's say you want to visualize "model.att4.query.weight" if it exists:
    key_to_plot = "model.att4.query.weight"
    if key_to_plot in state_dict:
        tensor = state_dict[key_to_plot]
        # Convert to numpy for plotting
        data = tensor.cpu().numpy().flatten()
        plt.hist(data, bins=50)
        plt.title(f"Histogram of {key_to_plot}")
        plt.xlabel("Value")
        plt.ylabel("Frequency")
        plt.show()
    else:
        print(f"{key_to_plot} not found in state_dict.")
```
